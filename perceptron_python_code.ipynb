```python
def step(value):
  return 1 if value>0 else 0
def perceptron(x,y):
  #firstly insert 1 with the x array
  x=np.insert(x,0,1,axis=1)
  w=np.ones(x.shape[1])
  ir=0.1
  for i in range(1000):
    print('loop---------------------',i)
    j=np.random.randint(0,100)
    # dot of x and w
    y_hat=step(np.dot(x[j],w))
    #now the final weight function
    w=w+(y[j]-y_hat)*ir*x[j]
  return w[0],w[1:]
int_,coeff=perceptron(x,y)
```

```python
print(coeff)
print(int_)
```

```python
m=-(coeff[0]/coeff[1])
b=-(int_/coeff[1])
m,b
```
