```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
#we are getting Training dataset for model
training_set=tf.keras.utils.image_dataset_from_directory(
"/kaggle/input/fruit-and-vegetable-image-recognition/train", # directory for images
    labels="inferred", #it will take labels name from directory's file name(apple,banana...)
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
#we are getting Validation dataset for model
validation_set=tf.keras.utils.image_dataset_from_directory(
"/kaggle/input/fruit-and-vegetable-image-recognition/validation", # directory for images
    labels="inferred", #it will take labels name from directory's file name(apple,banana.)
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
#build the cnn model
cnn=tf.keras.models.Sequential()
```

```python
#build the convolutional layer
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```python
#build the pooling layer and pool size is 2 mean 2x2 matrix
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#convolutional layers
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'))
#pooling layers
cnn.add(tf.keras.layers.MaxPool2D())
```

```python
# avoid overfitting
cnn.add(tf.keras.layers.Dropout(0.5))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
#now the fc layers
cnn.add(tf.keras.layers.Dense(units=128,activation="relu",)) # units is number of neuron number
```

```python
#output layer
cnn.add(tf.keras.layers.Dense(units=36,activation="softmax")) 
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
training=cnn.fit(x=training_set,validation_data=validation_set,epochs=32)
```

```python
# now evaluate the model
train_loss,train_acc=cnn.evaluate(training_set)
print(train_acc)
```

```python
#now check on the validation
val_loss,val_acc=cnn.evaluate(validation_set)
print(val_acc)
```

```python
#now save the model
cnn.save('trained_model.h5')
```
